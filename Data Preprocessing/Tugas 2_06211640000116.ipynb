```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("Outbreak_240817.csv")
data.head()
```

```python
dataa=data.drop('Id',1)
```

```python
dataa.describe()
```

```python
np.sum(data.isnull())
```

```python
data['sumAtRisk']=data['sumAtRisk'].fillna(data['sumAtRisk'].mean())
data['sumCases']=data['sumCases'].fillna(data['sumCases'].mean())
data['sumDeaths']=data['sumDeaths'].fillna(data['sumDeaths'].mean())
data['sumDestroyed']=data['sumDestroyed'].fillna(data['sumDestroyed'].mean())
data['sumSlaughtered']=data['sumSlaughtered'].fillna(data['sumSlaughtered'].mean())
```

```python
from scipy.stats import mode
data['observationDate']=data['observationDate'].fillna(data['observationDate']).dropna().mode()[0]
data['serotypes']=data['serotypes'].fillna(data['serotypes']).dropna().mode()[0]
data['speciesDescription']=data['speciesDescription'].fillna(data['speciesDescription']).dropna().mode()[0]
```

```python
np.sum(data.isnull())
```

```python
from scipy import stats
data1=data.drop(['observationDate','serotypes','speciesDescription'],1)
z=np.abs(stats.zscore(data1._get_numeric_data()))
print(z)
data=data[(z<3).all(axis=1)]
print(data.shape)
```
