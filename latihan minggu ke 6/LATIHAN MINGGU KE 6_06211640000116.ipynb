```python
import pandas as pd
import numpy as np
```

```python
sydney=pd.read_csv("weatherAUS.csv")
sydney.tail()
```

```python
sydney.describe()
```

```python
np.sum(sydney.isnull())
```

```python
sydney["MinTemp"]=sydney["MinTemp"].fillna(sydney["MinTemp"].mean())
sydney["MaxTemp"]=sydney["MaxTemp"].fillna(sydney["MaxTemp"].mean())
sydney["Rainfall"]=sydney["Rainfall"].fillna(sydney["Rainfall"].mean())
sydney["Evaporation"]=sydney["Evaporation"].fillna(sydney["Evaporation"].mean())
sydney["Sunshine"]=sydney["Sunshine"].fillna(sydney["Sunshine"].mean())
sydney["WindGustSpeed"]=sydney["WindGustSpeed"].fillna(sydney["WindGustSpeed"].mean())
sydney["WindSpeed9am"]=sydney["WindSpeed9am"].fillna(sydney["WindSpeed9am"].mean())
sydney["WindSpeed3pm"]=sydney["WindSpeed3pm"].fillna(sydney["WindSpeed3pm"].mean())
sydney["Humidity9am"]=sydney["Humidity9am"].fillna(sydney["Humidity9am"].mean())
sydney["Humidity3pm"]=sydney["Humidity3pm"].fillna(sydney["Humidity3pm"].mean())
sydney["Pressure9am"]=sydney["Pressure9am"].fillna(sydney["Pressure9am"].mean())
sydney["Pressure3pm"]=sydney["Pressure3pm"].fillna(sydney["Pressure3pm"].mean())
sydney["Cloud9am"]=sydney["Cloud9am"].fillna(sydney["Cloud9am"].mean())
sydney["Cloud3pm"]=sydney["Cloud3pm"].fillna(sydney["Cloud3pm"].mean())
sydney["Temp9am"]=sydney["Temp9am"].fillna(sydney["Temp9am"].mean())
sydney["Temp3pm"]=sydney["Temp3pm"].fillna(sydney["Temp3pm"].mean())
sydney["RISK_MM"]=sydney["RISK_MM"].fillna(sydney["RISK_MM"].mean())
```

```python
np.sum(sydney.isnull())
```

```python
from scipy.stats import mode
sydney['WindGustDir']=sydney['WindGustDir'].fillna(sydney['WindGustDir']).dropna().mode()[0]
sydney['WindDir9am']=sydney['WindDir9am'].fillna(sydney['WindDir9am']).dropna().mode()[0]
sydney['WindDir3pm']=sydney['WindDir3pm'].fillna(sydney['WindDir3pm']).dropna().mode()[0]
sydney['RainToday']=sydney['RainToday'].fillna(sydney['RainToday']).dropna().mode()[0]
sydney['RainTomorrow']=sydney['RainTomorrow'].fillna(sydney['RainTomorrow']).dropna().mode()[0]
```

```python
np.sum(sydney.isnull())
```

```python
from scipy import stats
sydney1=sydney.drop(['Date','Location','WindGustDir','WindDir9am','WindDir3pm','RainToday','RainTomorrow'],1)
z=np.abs(stats.zscore(sydney1._get_numeric_data()))
print(z)
sydney=sydney[(z<3).all(axis=1)]
print(sydney.shape)
```

```python
sydney2=sydney.drop(['Date','Location','WindGustDir','WindDir9am','WindDir3pm','RainToday','RainTomorrow'],1)
sydney2.shape
```

```python
#normalisasi
from sklearn.preprocessing import normalize
xnorm=normalize(sydney2,norm='l2') #default norm l2, other are l1 and max
xnorm=pd.DataFrame(xnorm)
xnorm.tail()
```

```python
from sklearn.preprocessing import LabelEncoder
kategori=['WindGustDir','WindDir9am','WindDir3pm','RainToday','RainTomorrow']
for feature in kategori:
    if feature in sydney.columns.values:
        sydney[feature]=LabelEncoder().fit_transform(sydney[feature])
sydney.tail()
```

```python
#feature selection (mencari skor)
from sklearn.feature_selection import SelectKBest, chi2, f_regression
x=sydney2.iloc[:,0:19]
y=sydney.iloc[:,-1]
selector=SelectKBest(f_regression, k="all").fit(x,y)
X_new=selector.transform(x)
scores=selector.scores_
scores
```
